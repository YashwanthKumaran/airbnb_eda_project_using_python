**Task 4: Data Cleaning**

```python
data.isnull().sum()
```

```python
#dropping all missing values
data.dropna(inplace = True)

#data.fillna() -- Used to replace missing (NaN) values in a DataFrame or Series with a specified value.
```

```python
data.isnull().sum().sum()
```

```python
#Handling duplicate rows in a dataset.
data.duplicated().sum()
data[data.duplicated()]
```

```python
#Deleting duplicate rows from the dataset.
data.drop_duplicates(inplace = True)
data.duplicated().sum()
```

```python
#Type Casting
#Converting data from one type to another
data["id"] = data["id"].astype("object")
data["host_id"] = data["host_id"].astype("object")
data.dtypes
```

**Task 5: Data Analysis**

**Univariate Analysis**

```python
#Identifying outliers in price
df = data[data["price"]<1500]
sns.boxplot(data = df,x = "price")
```

```python
#Analyzing prize distribution
plt.figure(figsize = (8,5))
sns.histplot(data = df,x = "price",bins = 100)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Price Distribution")
plt.show()
```

```python
plt.figure(figsize = (6,3))
sns.histplot(data = df , x = "availability_365")
plt.ylabel("Frequency")
plt.xlabel("Availability")
plt.title("Availability Distribution")
plt.show()
```

```python
 #Price per bed
 df["price_per_bed"] = df["price"] / df["beds"]
```

```python
#Average price per bed
df.groupby(by = "neighbourhood_group")["price_per_bed"].mean()
```

**Bivariate Analysis**

```python
df.columns
```

```python
#Price dependencies in neighbourhood
sns.barplot(data = df,x = "neighbourhood_group",y = "price_per_bed",hue = "room_type")
plt.show()
```

```python
plt.figure(figsize= (8,5))
plt.title("Locality and Review dependency")
sns.scatterplot(data = df, x = "number_of_reviews",y = "price", hue = "neighbourhood_group")
plt.show()
```

```python
sns.pairplot(data = df, vars = ["price","minimum_nights","number_of_reviews","availability_365"],hue = "room_type")
plt.show()
```

```python
#Geographical distribution of AirBnb listing
plt.figure(figsize = (8,5))
sns.scatterplot(data = df, x = "longitude",y = "latitude",hue = "room_type")
plt.title("Geographical distribution of AirBnb listing")
plt.show()
```
